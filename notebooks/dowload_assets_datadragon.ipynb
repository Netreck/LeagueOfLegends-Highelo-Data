```python
import os
import requests

# URL base do CommunityDragon para os ícones de campeões
BASE_URL = "https://raw.communitydragon.org/latest/plugins/rcp-be-lol-game-data/global/default/v1/champion-icons/"

# Criação da pasta para salvar os ícones
OUTPUT_DIR = "champion_icons"
os.makedirs(OUTPUT_DIR, exist_ok=True)

def download_champion_icons():
    # IDs dos campeões vão de 1 a 999 (arbitrário para cobrir todos os campeões)
    print("Baixando ícones de campeões...")
    for champ_id in range(1, 1000):
        url = f"{BASE_URL}{champ_id}.png"
        response = requests.get(url)

        # Verifica se o arquivo existe
        if response.status_code == 200:
            file_path = os.path.join(OUTPUT_DIR, f"{champ_id}.png")
            with open(file_path, "wb") as file:
                file.write(response.content)
            print(f"Baixado: {champ_id}.png")
        elif response.status_code == 404:
            # Para IDs inexistentes, ele simplesmente pula
            continue
        else:
            print(f"Erro ao baixar o ID {champ_id}: {response.status_code}")
    print("Download concluído!")

# Executar o download
if __name__ == "__main__":
    download_champion_icons()
```

```python
import requests
import os

champion_json = {}

# Função para obter a versão mais recente do dDragon
def get_latest_ddragon():
    global champion_json

    if champion_json:
        return champion_json

    # Obtendo a versão mais recente
    versions_response = requests.get("https://ddragon.leagueoflegends.com/api/versions.json")
    versions = versions_response.json()
    latest_version = versions[0]

    # Obtendo os campeões
    champions_response = requests.get(f"https://ddragon.leagueoflegends.com/cdn/{latest_version}/data/en_US/champion.json")
    data = champions_response.json()
    champions = data["data"]

    champion_json = champions
    return champions

# Função para obter um campeão pelo seu ID
def get_champion_by_key(key):
    champions = get_latest_ddragon()

    for champion_name, champion_data in champions.items():
        if champion_data["key"] == str(key):
            return champion_data

    return False

# Função para renomear os arquivos da pasta champion_icons
def rename_champion_icons(champion_json, folder_path="champion_icons"):
    if not os.path.exists(folder_path):
        print(f"Pasta {folder_path} não encontrada!")
        return

    # Renomeando os arquivos
    for champion_id, champion_data in champion_json.items():
        champion_name = champion_data["name"].lower().replace(" ", "_")  # Ajuste o nome do campeão para o formato adequado
        old_filename = f"{champion_data['key']}.png"
        new_filename = f"{champion_name}.png"
        
        old_path = os.path.join(folder_path, old_filename)
        new_path = os.path.join(folder_path, new_filename)
        
        # Verifica se o arquivo existe antes de renomear
        if os.path.exists(old_path):
            os.rename(old_path, new_path)
            print(f"Renomeado: {old_filename} -> {new_filename}")
        else:
            print(f"Arquivo {old_filename} não encontrado!")


champions = get_latest_ddragon()


rename_champion_icons(champions)
```

```python
import os
import aiohttp
import asyncio

# URL 
base_url = "https://raw.communitydragon.org/14.24/game/assets/ux/summonericons/"


os.makedirs('profileicon', exist_ok=True)

async def download_icon(session, icon_id):
    image_url = f"{base_url}profileicon{icon_id}.png"
    try:
        async with session.get(image_url) as response:
            if response.status == 200:  
                with open(f'profileicon/{icon_id}.png', 'wb') as f:
                    f.write(await response.read())
                print(f"Ícone {icon_id} baixado com sucesso!")
            else:
                print(f"Ícone {icon_id} não encontrado.")
    except Exception as e:
        print(f"Erro ao baixar o ícone {icon_id}: {e}")

async def main():
    async with aiohttp.ClientSession() as session:
        tasks = []
        for i in range(1, 6800):  #total icones
            tasks.append(download_icon(session, i))
        await asyncio.gather(*tasks)


await main()
```
